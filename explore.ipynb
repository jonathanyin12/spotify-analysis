# **Analyzing all songs in official Spotify playlists**

```python
from scrape import get_spotify_songs
import pandas as pd

# spotify_df = get_spotify_songs()
# spotify_df.to_csv('all_spotify_songs.csv',index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

spotify_df = pd.read_csv('all_spotify_songs.csv')
popular = spotify_df.loc[spotify_df['popularity'] >=90]
popular.sort_values('danceability').tail(15)
```

```python
correlation_df = spotify_df.loc[:, 'popularity':].corr()
plt.figure(figsize=(12, 12))
sns.heatmap(correlation_df, annot=True, square=True, cmap='Blues', vmax=1, vmin=-1)
plt.show()
```

# **Get all liked songs**

```python
from playlists import user_playlists, get_playlist_tracks_ids
from tracks import track_feature_cols, track_meta_cols, get_tracks_data
all_tracks = set()
for playlist in user_playlists:
    if playlist['owner']['display_name']=='Jonathan Yin':
        track_ids = get_playlist_tracks_ids(playlist['id'])
        all_tracks.update(track_ids)

tracks_data = get_tracks_data(list(all_tracks))
df = pd.DataFrame(tracks_data, columns=  track_meta_cols+track_feature_cols)
df
```

```python
all_liked = pd.concat([liked_tracks, df], ignore_index=True).drop_duplicates()
all_liked.sort_values('energy').tail(30)
```
