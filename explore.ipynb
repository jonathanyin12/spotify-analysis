# **Analyzing all songs in official Spotify playlists**

```python
from scrape import get_spotify_songs
import pandas as pd

# spotify_df = get_spotify_songs()
# spotify_df.to_csv('all_spotify_songs.csv',index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

spotify_df = pd.read_csv('all_spotify_songs.csv')
popular = spotify_df.loc[spotify_df['popularity'] >=90]
popular.sort_values('danceability').tail(15)
```

```python
correlation_df = spotify_df.loc[:, 'popularity':].corr()
plt.figure(figsize=(12, 12))
sns.heatmap(correlation_df, annot=True, square=True, cmap='Blues', vmax=1, vmin=-1)
plt.show()
```
